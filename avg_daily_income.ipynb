```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt 
import seaborn as sns
```

```python
avg_daily_income_full = pd.read_csv('./data/avg_daily_income_full.csv')
avg_daily_income_full
```

```python
avg_daily_income_country = avg_daily_income_full[['continent','country','country_code']]
avg_daily_income_country
```

```python
avg_daily_income_num = avg_daily_income_full.drop(['continent','country','country_code'], axis=1)
avg_daily_income_num
```

```python
avg_daily_income = avg_daily_income_num.applymap(lambda x: x / 1000 if x > 100 else x)
avg_daily_income
```

```python
avg_daily_income = pd.concat([avg_daily_income_country,avg_daily_income], axis=1)
avg_daily_income
```

```python
avg_daily_income.to_csv('./data/avg_daily_income_corrected.csv', index=False)
```
